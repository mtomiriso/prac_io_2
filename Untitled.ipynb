```python
import scipy as sc
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from sklearn.metrics import r2_score
import statsmodels.api as sm
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.tsa.arima.model import ARIMA


%matplotlib inline
```

```python
data = pd.read_csv('training.csv')
ts = pd.Series(np.asarray(data['Value'], dtype=np.float64), index=data['Date'])
ts.cumsum()
ts.plot(figsize=(20, 8))
```

**plot moving stats**

```python
def plotStats(ts):
    ts.plot(color='b', legend=True, figsize=(20, 8), label='original row')
    ts.rolling(window=12,center=False).mean().plot(color='g', legend=True, label='mean')
    ts.rolling(window=12, center=False).std().plot(color='r', legend=True, label = 'std')
    
plotStats(ts)
```

**DF test**

```python
def dfTest(ts, isPrint=True):
    test = sm.tsa.adfuller(ts)
    if isPrint:
        print ('adf: ', test[0]) 
        print ('p-value: ', test[1])
        print ('Critical values: ', test[4])
    if test[0]> test[4]['5%']: 
        if isPrint:
            print ('have unit roots, a number is not stationary')
        return False
    else:
        if isPrint:
            print ('unit roots there, a number of stationary')
        return True
    return False

dfTest(ts)
```

**decomposition into trend, residue, sezonalnost**

```python
data.reset_index(inplace=True)
data['Date'] = pd.to_datetime(data['Date'])
data = data.set_index('Date')
decomp = seasonal_decompose(data.Value)
def showPlot(x, title=''):
    plt.figure(figsize=(20, 4))
    plt.plot(x)
    plt.title(title)
    plt.show()

trend = decomp.trend
resid = decomp.resid
seasonal = decomp.seasonal

showPlot(trend, 'Trend')
showPlot(resid, 'Resid')
showPlot(seasonal, 'Seasonal')
```

**check the trend stationarity**


```python
print (trend.keys())
trend.dropna(inplace=True)
tsTrend = pd.Series(np.asarray(trend, dtype=np.float64), index=trend.index)
plotStats(tsTrend)

dfTest(tsTrend)
```

**check the resid stationarity**


```python
resid.dropna(inplace=True)
tsResid = pd.Series(np.asarray(resid, dtype=np.float64), index=resid.index)
tsResid.cumsum()
plotStats(tsResid)
dfTest(tsResid)
```

**check the seasonal stationarity**


```python
seasonal.dropna(inplace=True)
tsSeasonal = pd.Series(np.asarray(seasonal, dtype=np.float64), index=seasonal.index)
tsSeasonal.cumsum()
plotStats(tsSeasonal)
dfTest(tsSeasonal)
```

**Find order of integration**

```python
def findOrder(ts, isPrint=False):
    integrationOrder = 0
    while not dfTest(ts, isPrint):
        ts = ts.diff(periods=1).dropna()
        integrationOrder += 1
    return integrationOrder
```

```python
k = findOrder(ts)
print (k)
```

**Find autocorrelation function parameters**


```python
log_shift_ts = (np.log(ts) - np.log(ts).shift())
log_shift_ts.dropna(inplace=True)
acf_values = acf(log_shift_ts)
pacf_values = pacf(log_shift_ts, method='ols')
    
    
plt.figure(figsize=(10, 8))    
plt.plot(acf_values)
plt.title('ACF')
plt.show()

plt.figure(figsize=(10, 8))
plt.plot(pacf_values)
plt.title('PACF')
plt.show()
```

**Train ARIMA model**



```python
ts.index = pd.to_datetime(ts.index)
ts_log = np.log(ts)
model = ARIMA(ts_log, order=(int(pacf_values[0]), k, int(acf_values[0])))
fit_val = model.fit(show = False)
pred = fit_val.fittedvalues

ts_log_diff = (np.log(ts) - np.log(ts).shift())
ts_log_diff.dropna(inplace=True)
pred.plot(figsize=(20, 8), color='r')
ts_log_diff.plot(figsize=(20, 8), color='b')
print ('R^2 value: ', r2_score(ts_log_diff, pred))
```

**Check on test set**



```python
dataTest = pd.read_csv('testing.csv')
tsTest = pd.Series(np.asarray(dataTest['Value'], dtype=np.float64), index=dataTest['Date'])
tsTest.cumsum()

predict = fit_val.predict('1988-12-01', '1993-12-01')
predict = predict['1989-01-01':]
```

```python
predict[0] += np.log(tsTest[0])
predict = predict.cumsum()
predict = np.exp(predict) 
plt.figure(figsize=(12,8))
plt.plot(predict, color='r')
plt.plot(tsTest, color='b')
print ('R2 Score: %.4f' % r2_score(predict, tsTest))
```
